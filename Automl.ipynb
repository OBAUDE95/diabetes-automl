```python
from pycaret.classification import *

clf_setup = setup(
    data=df,
    target='Outcome',
    session_id=42,
    normalize=True,
    verbose=False    # suppress extra logs; no need for 'silent'
)
```


clf = setup(
    data=df,
    target='Outcome',
    session_id=42,
    train_size=0.8,
    normalize=True,
    remove_outliers=True,
    imputation_type='simple',
    numeric_imputation='mean',
    categorical_imputation='mode',
    silent=True
)


```python
best_model = compare_models()
```

```python
tuned_model = tune_model(best_model, optimize='F1', n_iter=30)
```

```python
predictions = predict_model(tuned_model)
predictions.head()
```
